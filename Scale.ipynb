```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
nodes = [2,4,6,8,10,12]
mpi = 1.76/np.array([1.76,1.64,1.35,1.16,1.04,0.94])
openmpi = 1.08/np.array([1.08,1.02,0.79,0.59,0.54,0.49])
nodes_ = [4,6,8,10,12]
hybrid = .88/np.array([0.88,1.03,0.8,0.73,.64])
```

```python
plt.scatter(nodes,mpi)
plt.title('MPI')
```

```python
plt.title('OPEN MPI')
plt.scatter(nodes,openmpi)
```

```python
plt.title('Hybrid')
plt.scatter(nodes_,hybrid)
```

```python
weaknodes = [4,6,8,10,12]
weakmpi = .52/np.array([.52,.50,.57,.64,.8])
weakopen = .11/np.array([0.11,.23,.3,.49,0.94])
weakhybrid = .6/np.array([.6,.73,.8,.85,0.81])
```

```python
plt.scatter(weaknodes,weakmpi)
plt.title('WEAK MPI')
```

```python
plt.title('WEAK OPEN MPI')
plt.scatter(weaknodes,weakopen)
```

```python
plt.title('WEAK Hybrid')
plt.scatter(weaknodes,weakhybrid)
```
